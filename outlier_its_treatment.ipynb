```python
salary = [11, 40, 45, 68, 65, 68, 78, 90, 57, 74, 91, 92, 88, 68, 57, 48, 99, 101, 68, 77, 110, 140]
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
# five point summary --> min, 25%, 50% , 75% ,max
```

```python
np.quantile(salary, [0, .25, .50, .75, 1])
```

```python
df = pd.DataFrame(salary, columns = ['Salary'])
```

```python
df.describe()
```

```python
# how to check ouutlier --. boxplot, distplot, 
```

```python
plt.figure(figsize = (12, 5))
plt.subplot(1,2,1)
sns.histplot(df['Salary'], kde=True)
plt.title("Distribution plot")

plt.subplot(1,2,2)
sns.boxplot(data = df, x= 'Salary')
plt.title("Box plot")

plt.show()
```

how to outlier 
1. dropping the outlier
2. capping the outlier
3. replace with the mean and median
4. scaling and transformation

```python
Q1 = df["Salary"].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
```

```python
Q1
```

```python
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR
```

```python
lower_fence
```

```python
upper_fence
```

```python
df.shape
```

```python
# dropping the outlier

df_filtered = df[(df.Salary >= lower_fence) & (df.Salary <= upper_fence)]
```

```python
df_filtered
```

```python
df_filtered.shape
```

```python
plt.figure(figsize = (12, 5))
plt.subplot(1,2,1)
sns.histplot(df_filtered['Salary'], kde=True)
plt.title("Distribution plot")

plt.subplot(1,2,2)
sns.boxplot(data = df_filtered, x= 'Salary')
plt.title("Box plot")

plt.show()
```

```python
# replace the outlier with mean and median
df['Salary_mean_imputed'] = np.where((df.Salary >= upper_fence) | (df.Salary <= lower_fence), df['Salary'].mean(), df['Salary'])
```

```python
df
```

```python
plt.figure(figsize = (12, 5))
plt.subplot(1,2,1)
sns.histplot(df['Salary_mean_imputed'], kde=True)
plt.title("Distribution plot")

plt.subplot(1,2,2)
sns.boxplot(data = df, x= 'Salary_mean_imputed')
plt.title("Box plot")

plt.show()
```

```python
# capping the outlier --> replacing outlier with the nearest values that is not outlier
```

```python
lower_cap = df['Salary'].quantile(0.05)     #lower cap with 5% percentile
upper_cap = df['Salary'].quantile(0.95)
```

```python
lower_cap
```

```python
upper_cap
```

```python
df['Salary_capped'] = np.where(df['Salary']< lower_cap, lower_cap,
                              np.where(df['Salary']> upper_cap, upper_cap, df['Salary']))
```

```python
df
```

```python
plt.figure(figsize = (12, 5))
plt.subplot(1,2,1)
sns.histplot(df['Salary_capped'], kde=True)
plt.title("Distribution plot")

plt.subplot(1,2,2)
sns.boxplot(data = df, x= 'Salary_capped')
plt.title("Box plot")

plt.show()
```
